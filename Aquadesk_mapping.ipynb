Voor beide categorieën is er een poging gedaan om de kolomnamen te mappen naar DarwinCore velden: https://dwc.tdwg.org/terms/. In het geval van Aquadesk gaat het uitsluitend om aantallen (of dichtheden) van biologische waarnemingen, waardoor het dus hoofdzakelijk onder de Occurrence type record valt met eventueel wat Event-metadata erbij waarvoor eerder de belangrijkste Darwin Core elementen zijn genoteerd. Dit betekent dat een DarwinCore Aquadesk dataset (.zip) als volgt kan worden ingedeeld:
- event.txt (waar en wanneer is het monster genomen)
- occurrence.txt (wat is er gemeten)
- meta.xml
- eml.xml

Om tot deze DarwinCore Aquadesk datasets te kunnen komen, moeten we onderzoeken of we voor de event en occurrence type record de “required” velden dekken. Als we de DarwinCore elementen van de bovenstaande tabellen filteren voor “Event” en “Occurrence” komen we op de volgende elementen uit als "Essentieel".

Event
- eventDate, eventID, decimalLongitude, decimalLatitude

Occurrence
- eventDate, decimalLongitude, decimalLatitude, occurrenceID, scientificName, occurrenceStatus, basisOfRecord

Om te kijken of we deze DarwinCore elementen kunnen mappen aan de Aquadesk data, is er een poging gedaan om alle kolomnamen te mappen naar relevante DarwinCore elementen. Onderstaande tabellen geven voor een deel van de kolommen voor Phytoplankton en Zoöplankton een eerste draft van mogelijke DarwinCore kolomnamen. Dit moet nog verder worden onderzocht en zal nog veel aanpassingen omvatten.


```python
aquadesk_to_dwc = {
    "Supplier": "recordedBy",
    "Projects": "datasetID",
    "Code": "occurrenceID",
    "Description": "locality",
    "Parameter": "scientificName",
    "LifeForm": "lifeStage",
    "Value": "organismQuantity",
    "Count": "organismQuantity2",
    "Uom": "measurementUnit",
    "Type": "measurementType",
    "ResultOf": "basisOfRecord",
    'SamplingMethod' : 'samplingProtocol',
    'AnalysisMethod': "measurementMethod",
    'Compartment_explanation' : 'habitat',
    "ResultTime": "eventDate",  
    "Quantity_explanation": "organismQuantityType",
    "TaxonType": "organismRemarks",
    "TaxonGroup": "taxonRemarks",
    "TaxonRank": "taxonRank",
    "MeasurementPackage": "measurementRemarks",
    "MonitoringNetworks": "locationRemarks",
    "license": "license",
    "language": "language",
    "longitude": "decimalLongitude",
    "latitude": "decimalLatitude"
}

dwc_columns = list(aquadesk_to_dwc.values())
df_dwc = df_aquadesk.rename(columns=aquadesk_to_dwc)[dwc_columns]

column_order = [
    "eventDate",
    "decimalLongitude",
    "decimalLatitude",
    "scientificName",
    "organismQuantity2",
    "basisOfRecord",
    "datasetID",
    "occurrenceID",
    "recordedBy",
    "organismQuantity",
    "measurementUnit",
    "measurementType",
    "samplingProtocol",
    "measurementMethod",
    "habitat",
    "lifeStage",
    "organismQuantityType",
    "organismRemarks",
    "taxonRemarks",
    "taxonRank",
    "measurementRemarks",
    "locationRemarks",
    "locality",
    "license",
    "language"
]

df_dwc = df_dwc[column_order]
df_dwc
```

```python
df_FYTPT = df_dwc[df_dwc['organismRemarks'] == 'FYTPT']
df_FYTPT = df_FYTPT.reset_index(drop=True)
df_FYTPT
```

Als we kijken naar de essentiële DarwinCore elementen, zien we dat deze (bijna) allemaal gevuld kunnen worden door de Aquadesk data als volgt:
- eventDate:
Gegeven door de “ResultTime” kolom, e.g. 2022-12-31T23:00:00Z.
- eventID:
Niet expliciet aanwezig, zou kunnen worden gegenereerd door bijvoorbeeld kolommen samen te nemen zoals de “Organisation”, “Projects” en “Code”. 
- decimalLongitude & decimalLatitude:
Beschikbaar binnen “GeoGraphy”, die Polygonen en MultiPolygonen bevat. Wij hebben het middelpunt genomen van deze Polygonen voor elke meting als Longitude en Latitude waarde.
- occurrenceID:
Niet expliciet aanwezig, zou kunnen worden gegenereerd door bijvoorbeeld kolommen samen te nemen zoals de “Organisation”, “Longitude”, “Latitude”, “ResultTime” en “Parameter”. 
- scientificName:
Gegeven door de “Parameter” kolom, e.g. Phalacrocorax carbo.
- occurrenceStatus:
De noodzakelijke identificatie in de vorm van presence of absence kan worden afgeleid van het aantal gemeten beestjes via de “count” kolom.
- basisOfRecord:
Gegeven door de “ResultOf” kolom, e.g. observations, extrapolation, observed.
