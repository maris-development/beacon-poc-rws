```python
from beacon_api import *
import pandas as pd
```

```python
client = Client("https://beacon-ihm.maris.nl")
```

```python
tables = client.list_tables()
tables['ihm_wadar'].get_table_schema()
```

```python
df_wadar = (
    tables["ihm_wadar"]
    .query()
    .add_select_column("Loc.Lon")
    .add_select_column("Loc.Lat")
    .add_select_column("Tijdstip")
    .add_select_column("Waarde_Numeriek")
    .add_select_column("Locatie")
    .add_select_column("Statuswaarde")
    .add_select_column("Bemonsteringshoogte")
    .add_select_column("Referentievlak")
    .add_select_column("OpdrachtgevendeInstantie")
    .add_select_column("Kwaliteitswaardecode")
    .add_select_column("Loc.Code")
    .add_select_column("Meta.Compartiment.Code")
    .add_select_column("Meta.Grootheid.Code")
    .add_select_column("Meta.Grootheid.Oms")
    .add_select_column("Meta.Parameter.Code")
    .add_select_column("Meta.Eenheid")
    .add_select_column("Meta.Hoedanigheid")
    .add_select_column("Meta.WaardeBewerkingsMethode")
    .add_filter(EqualsFilter("Locatie", eq = 'Ameland, Nes (Ameland, Nes)'))
    .to_pandas_dataframe()
)
```

```python
df_wadar['Meta.Parameter.Code'] = df_wadar['Meta.Parameter.Code'].replace("NVT", None)
df_wadar
```

Om te kunnen voldoen aan de belangrijkste ODV output uitdaging is er geprobeerd om de relevante SeaDataNet Vocabulary P01 voor parameters en P06 voor units te koppelen aan elke meting. Dit is gedaan door te kijken naar alle unieke combinaties van de volgende kolommen:
- Meta.Grootheid.Code → Windsnelheid, waterhoogte, etc.
- Meta.Parameter.Code → Chemische stof, alleen ingevuld voor Meta.Grootheid.Code Concentratie
- Meta.Hoedanigheid → Referentievlak
- Meta.Compartiment.Code → Oppervlaktewater, lucht, etc.
- Meta.Eenheid → Graad, centimeter
- Meta.Parameter_Wat_omschrijving → Windsnelheid in Lucht in m/s, etc.


```python
unique_combinations = df_wadar_metadata[['Meta.Grootheid.Code', 
                                         'Meta.Parameter.Code', 
                                         'Meta.Hoedanigheid', 
                                         'Meta.Compartiment.Code',
                                         'Meta.Eenheid',
                                         'Meta.Parameter_Wat_Omschrijving']].drop_duplicates()

unique_combinations = unique_combinations.sort_values(by=['Meta.Grootheid.Code', 'Meta.Parameter.Code'], ascending=False)
unique_combinations = unique_combinations.reset_index(drop=True)
unique_combinations
```

Linking P01 and P06 to the WADAR data rows

```python
key_cols = [
    "Meta.Grootheid.Code",
    "Meta.Parameter.Code",
    "Meta.Hoedanigheid",
    "Meta.Compartiment.Code",
    "Meta.Eenheid"
]

value_cols = ["P01 mapping", "P06 mapping"]

unique_combinations_filtered = unique_combinations[key_cols + value_cols].drop_duplicates().reset_index(drop=True)

merged_df = df_wadar.merge(
    unique_combinations_filtered[key_cols + value_cols],
    on=key_cols,
    how="left"
)
```
